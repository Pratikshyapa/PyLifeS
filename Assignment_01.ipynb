```python
import matplotlib.pyplot as plt
print()
X_Point=np.zeros(264)
Y_Point=np.zeros(264)
for i in range(264):
    X_Point[i]=float(A1[i])
    Y_Point[i]=float(B1[i])
print(X_Point)
print(Y_Point)

plt.scatter(X_Point,Y_Point)
plt.show()
```

```python
X2 = np.log(Y_Point)
X1 = np.log(X_Point)

# Import math Library
import math
from sklearn.linear_model import LinearRegression

X1 = X1.reshape(-1, 1) # make it a column vector

# Create a linear regression model
model = LinearRegression() 

# Fit the model to the data
model.fit(X1, X2)

# Get the coefficients of the fitted line
B = model.coef_[0] # slope
A1 = model.intercept_ # intercept

# Predict the values using the fitted model
predicted_X2 = model.predict(X1)

# Plot the data and fitted line
plt.scatter(X1,X2)
plt.plot(X1, predicted_X2, label=f'Fitted line: f(t) = {B:.2f}x + {A1:.2f}', color='red')
plt.xlabel('Body weight',fontsize=20)
plt.ylabel('BMR',fontsize=20)
plt.title('Linear Regression for BMR',fontsize=20)
plt.legend()
plt.savefig("BMR_BW.png")
plt.show()

print(f"Slope of the fitted line: {B:.2f}")
print(f"Intercept of the fitted line: {A1:.2f}")
print("A =", math.e**A1)
print(f"B = {B:.2f}")
```
