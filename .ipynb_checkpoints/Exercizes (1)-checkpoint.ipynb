```python
A=np.loadtxt('YodasLaw.csv', delimiter=",", dtype=str)
print(A)
```

```python
X_Point=np.zeros(99)
Y_Point=np.zeros(99)
```

```python
for i in range(99):
    X_Point[i]=float(A1[i])
    Y_Point[i]=float(B1[i])
print(X_Point)
print(Y_Point)
import numpy as np
B=np.array(X_Point)

```

```python
plt.scatter(X_Point,Y_Point)
plt.show()

# Import curve fitting package from scipy
from scipy.optimize import curve_fit
def BMR_FIT(X1, A, B):
    return A * (X1**B)
params, covariance = curve_fit(BMR_FIT, X_Point, Y_Point)
# Fitted parameters
A_fit, B_fit = params
plt.plot(X1, BMR_FIT(X1, A, B), label=f'Fitted curve; A={A_fit:.2f}, B={B_fit:.2f}', color='red')
```

```python
print(f"Fitted A: {A_fit:.2f}")
print(f"Fitted B: {B_fit:.2f}")
```

```python
A=np.loadtxt('elgar.csv', delimiter=",", dtype=str)
print(A)
```

```python
A1=A[:,3]
B1=A[:,4]

print(len(B1))
print(len(A1))
```

```python
A1=A1[2:]
B1=B1[2:]
print(A1)
print(B1)
```

```python
X_Point=np.zeros(264)
Y_Point=np.zeros(264)
for i in range(264):
    X_Point[i]=float(A1[i])
    Y_Point[i]=float(B1[i])
print(X_Point)
print(Y_Point)

plt.scatter(X_Point,Y_Point)
plt.show()
```

```python
X2 = np.log(Y_Point)
X1 = np.log(X_Point)

# Import math Library
import math
from sklearn.linear_model import LinearRegression

X1 = X1.reshape(-1, 1) # make it a column vector

# Create a linear regression model
model = LinearRegression() 

# Fit the model to the data
model.fit(X1, X2)

# Get the coefficients of the fitted line
B = model.coef_[0] # slope
A1 = model.intercept_ # intercept

# Predict the values using the fitted model
predicted_X2 = model.predict(X1)

# Plot the data and fitted line
plt.scatter(X1,X2)
plt.plot(X1, predicted_X2, label=f'Fitted line: f(t) = {B:.2f}x + {A1:.2f}', color='red')
plt.xlabel('Body weight',fontsize=20)
plt.ylabel('BMR',fontsize=20)
plt.title('Linear Regression for BMR',fontsize=20)
plt.legend()
plt.savefig("BMR_BW.png")
plt.show()

print(f"Slope of the fitted line: {B:.2f}")
print(f"Intercept of the fitted line: {A1:.2f}")
print("A =", math.e**A1)
print(f"B = {B:.2f}")
```

```python
C=np.loadtxt('BMR_1.csv', delimiter=",", dtype=str)
print(C)
C1=C[:,4]
C2=C[:,5]
C1=C1[1:]
C2=C2[1:]
print(C1)
print(len(C1))
X_Point=np.zeros(70)
Y_Point=np.zeros(70)
for i in range(70):
    X_Point[i]=float(C1[i])
    Y_Point[i]=float(C2[i])
print(X_Point)
print(Y_Point)

from numpy import mean 
from numpy import std
from numpy import cov
print( 'data1: mean=%.3f stdv=%.3f' % (mean(X_Point), std(X_Point)))
print( 'data1: mean=%.3f stdv=%.3f' % (mean(Y_Point), std(Y_Point)))
covariance = cov(X_Point, Y_Point)
print(covariance)
```
